Using your knowledge of PySpark, Pandas, or SQL, you’ll determine if there is any bias towards reviews that were written as part of the Vine program. For this analysis, you'll determine if having a paid Vine review makes a difference in the percentage of 5-star reviews.

```python
import pandas as pd
import numpy
```

```python
# Read in vine table 

vine_df = pd.read_csv('./vine_table.csv') 
vine_df.head()
```

```python
vine_df.info()
```

```python
# 1
# Filter the data and create a new DataFrame or table to retrieve 
# all the rows where the total_votes count is equal to or greater 
# than 20 to pick reviews that are more likely to be helpful and 
# to avoid having division by zero errors later on.

vine_df_20 = vine_df.loc[vine_df['total_votes']>20]
vine_df_20
```

```python
#2
# Filter the new DataFrame or table created in Step 1 and 
# create a new DataFrame or table to retrieve all the rows where the 
# number of helpful_votes divided by total_votes is equal to or greater than 50%.
vine_df_cleaned = vine_df_20.loc[(vine_df_20['helpful_votes'] / vine_df_20['total_votes'])>=0.5]
vine_df_cleaned
```

```python
# 3
# Filter the DataFrame or table created in Step 2, 
# and create a new DataFrame or table that retrieves 
# all the rows where a review was written as part of 
# the Vine program (paid), vine == 'Y'.
paid_reviews_df = vine_df_cleaned.loc[vine_df_cleaned['vine'] =='Y']

paid_reviews_df
```

```python
# 4
# Repeat Step 3, but this time retrieve all the rows where the review was 
# not part of the Vine program (unpaid), vine == 'N'.
unpaid_reviews_df = vine_df_cleaned.loc[vine_df_cleaned['vine'] =='N']
unpaid_reviews_df

```

5.
Determine the total number of reviews, 
the number of 5-star reviews, 
and the percentage of 5-star reviews 
for the two types of review (paid vs unpaid).

```python
# Total number of reviews 
total_review_count = vine_df_cleaned['review_id'].count()
total_review_count
```

```python
# Paid reviews 
paid_reviews_count = paid_reviews_df['review_id'].count()
paid_reviews_count
```

```python
# Unpaid Reviews
unpaid_reviews_count = unpaid_reviews_df['review_id'].count()
unpaid_reviews_count
```

```python
# Total number of number of 5-star reviews
five_star_reviews = vine_df_cleaned.loc[vine_df_cleaned['star_rating']==5]['star_rating'].count()
five_star_reviews
```

```python
# Total Paid 5-Star Reviews 
paid_five_star_reviews = paid_reviews_df.loc[paid_reviews_df['star_rating'] == 5]['star_rating'].count() 
paid_five_star_reviews
```

```python
# Total Unpaid 5-Star Reviews 
unpaid_five_star_reviews = unpaid_reviews_df.loc[unpaid_reviews_df['star_rating'] == 5]['star_rating'].count() 
unpaid_five_star_reviews
```

```python
# The percentage of paid 5-star reviews
percent_paid_five_star_reviews = paid_five_star_reviews/ five_star_reviews *100
percent_paid_five_star_reviews
```

```python
# The percentage of unpaid 5-star reviews
percent_unpaid_five_star_reviews  = unpaid_five_star_reviews/ five_star_reviews *100
percent_unpaid_five_star_reviews
```
